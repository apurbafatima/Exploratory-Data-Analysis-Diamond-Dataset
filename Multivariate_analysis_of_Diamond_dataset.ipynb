```python
# Step 1: Check for missing values
missing_values = diamond_data.isnull().sum()

# Step 2: Data types - ensure numerical columns are correctly typed
# Converting numerical columns as appropriate (e.g., price, carat, depth, table, x, y, z)
diamond_data["price"] = pd.to_numeric(diamond_data["price"], errors='coerce')
diamond_data["carat"] = pd.to_numeric(diamond_data["carat"], errors='coerce')
diamond_data["depth"] = pd.to_numeric(diamond_data["depth"], errors='coerce')
diamond_data["table"] = pd.to_numeric(diamond_data["table"], errors='coerce')
diamond_data["x"] = pd.to_numeric(diamond_data["x"], errors='coerce')
diamond_data["y"] = pd.to_numeric(diamond_data["y"], errors='coerce')
diamond_data["z"] = pd.to_numeric(diamond_data["z"], errors='coerce')

# Step 3: Remove duplicates
diamond_data = diamond_data.drop_duplicates()

# Step 4: Standardize categorical data - Ensure consistency in category names for clarity, cut, color
diamond_data["clarity"] = diamond_data["clarity"].str.strip().str.upper()  # Ensure uniformity in clarity labels
diamond_data["cut"] = diamond_data["cut"].str.strip().str.title()           # Ensure consistency in cut names
diamond_data["color"] = diamond_data["color"].str.strip().str.upper()       # Ensure uniformity in color labels

# Step 5: Outlier detection - Here, I'll just identify outliers for review in key numerical columns.
# Checking for potential outliers in carat and price using a simple statistical approach (e.g., IQR)

# Calculate IQR for 'carat' and 'price' to detect outliers
Q1_carat = diamond_data['carat'].quantile(0.25)
Q3_carat = diamond_data['carat'].quantile(0.75)
IQR_carat = Q3_carat - Q1_carat

Q1_price = diamond_data['price'].quantile(0.25)
Q3_price = diamond_data['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price

# Filtering out potential outliers beyond 1.5 * IQR for carat and price
diamond_data = diamond_data[
    ~((diamond_data['carat'] < (Q1_carat - 1.5 * IQR_carat)) | (diamond_data['carat'] > (Q3_carat + 1.5 * IQR_carat)))
]
diamond_data = diamond_data[
    ~((diamond_data['price'] < (Q1_price - 1.5 * IQR_price)) | (diamond_data['price'] > (Q3_price + 1.5 * IQR_price)))
]

# Display cleaned data information
cleaned_data_info = diamond_data.info()
cleaned_data_head = diamond_data.head()

cleaned_data_info, cleaned_data_head
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/DiamondData-1.csv'  # Adjust this path as needed
diamond_data = pd.read_csv(file_path)
# Univariate Analysis: Numerical Features
plt.figure(figsize=(12, 8))

# Price Distribution
plt.subplot(2, 2, 1)
sns.histplot(diamond_data['price'], bins=30, kde=True)
plt.title('Price Distribution')

# Carat Distribution
plt.subplot(2, 2, 2)
sns.histplot(diamond_data['carat'], bins=30, kde=True)
plt.title('Carat Distribution')

# Depth Distribution
plt.subplot(2, 2, 3)
sns.histplot(diamond_data['depth'], bins=30, kde=True)
plt.title('Depth Distribution')

# Table Distribution
plt.subplot(2, 2, 4)
sns.histplot(diamond_data['table'], bins=30, kde=True)
plt.title('Table Distribution')
```

```python
# Multivariate Analysis: Carat vs. Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diamond_data, x='carat', y='price', alpha=0.6)
plt.title("Carat vs Price")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show()

# Multivariate Analysis: Price by Cut, Color, and Clarity
plt.figure(figsize=(15, 6))

# Price vs Cut
plt.subplot(1, 3, 1)
sns.boxplot(data=diamond_data, x='cut', y='price')
plt.title("Price by Cut")

# Price vs Color
plt.subplot(1, 3, 2)
sns.boxplot(data=diamond_data, x='color', y='price')
plt.title("Price by Color")

# Price vs Clarity
plt.subplot(1, 3, 3)
sns.boxplot(data=diamond_data, x='clarity', y='price')
plt.title("Price by Clarity")

plt.tight_layout()
plt.show()

# Pairplot to observe relationships between numerical features
sns.pairplot(diamond_data[["carat", "price", "depth", "table", "x", "y", "z"]])
plt.show()
```
